# Step_1:
Finding mean and standard deviation of data samples 

```python
#function to find sum
def sum_values(giveme):
  sum_i = 0
  for i in giveme:
    sum_i = sum_i + i;
  return sum_i
```

```python
fa_sum = sum_values(feature_A)
print("sum of feature fA: %.2f" %fa_sum)
fb_sum = sum_values(feature_B)
print("sum of feature fB: %.2f" %fb_sum)
fc_sum = sum_values(feature_C)
print("sum of feature fC: %.2f" %fc_sum)
```

```python
#creating a function to find mean values
N = len(feature_A)
def mean_me (N, sum_val):
  mean_values = 0
  mean_values = sum_val / N
  return mean_values
  
```

```python
#Finding mean values
fa_mean = mean_me(N,fa_sum)
print("mean of feature fA: %.2f" %fa_mean)
fb_mean = mean_me(N,fb_sum)
print("mean of feature fb: %.2f" %fb_mean)
fc_mean = mean_me(N,fc_sum)
print("mean of feature fc: %.2f" %fc_mean)
```

```python
fa_sum = sum_values(feature_A)
print("sum of featur fa: %.2f" %fa_sum)
```

```python
#Finding standard deviation
f_i = 0
std = 0
fi_sq = 0
def standard_deviation(N, fec, mean):
  fa_sq = 0
  for i in fec:
    f_i = (i - mean)
    fi_sq = (f_i * f_i)
    fa_sq += fi_sq
  std = (fa_sq / N )
  return std

```

```python
fa_std_Dev = standard_deviation(N, feature_A, fa_mean)
print("std of Feature_a: %.2f" %fa_std_Dev)

fb_std_Dev = standard_deviation(N, feature_A, fb_mean)
print("std of Feature_b: %.2f" %fb_std_Dev)

fc_std_Dev = standard_deviation(N, feature_A, fc_mean)
print("std of Feature_c: %.2f" %fc_std_Dev)
```

Making all features to have mean = 0 and std_dev = 1

# 2: Standardization

Formula:
stdise = (feature-mean) / std_Dev

```python
def standardise_me(fec, mean, std):
  coef  = 0
  coef_sum = 0 
  temp= 0
  stdise = 0
  for i in fec:
    coef = i - mean
    #coef_sum +=  coef
  temp += (coef/ std)
  stdise += temp

  return stdise


```

```python
stdise_fa = standardise_me(feature_A, fa_mean, fa_std_Dev)
print("standardise feature of fa: %.2f" %stdise_fa)

stdise_fb = standardise_me(feature_A, fa_mean, fb_std_Dev)
print("standardise feature of fb: %.2f" %stdise_fb)

stdise_fc = standardise_me(feature_A, fa_mean, fc_std_Dev)
print("standardise feature of fc: %.2f" %stdise_fc)
```

# **Calculating Co-variance for all features,** 

ie finding the linear relationship between features

variance = (feature - mean)^2 / N

co-_rariance = ((f_a - mean_a)(feature_b - mean_b)) / N



```python
#finding covariance first which is the spread of a data set around its mean value

def Co_variance (feature_1 , feature_2 , n, mean_1 , mean_2):
  temp1 = 0
  temp2 = 0
  corar = 0
  at_1 = 0
  at_2= 0


  for i in feature_1:
    at_1 = (i - mean_1)
    temp1 += at_1
  for j in feature_2:
    at_2 = (j - mean_2)
    temp2 += at_2

  covar = (temp1 * temp2) / (n - 1)

  return covar
```

```python
print("Covariance of feature_A & feature_A is =" ,Co_variance(feature_A , feature_A , N, fa_mean , fa_mean))
print("Covariance of feature_A & feature_B is =" ,Co_variance(feature_A , feature_B , N, fa_mean , fb_mean))
print("Covariance of feature_A & feature_c is =" ,Co_variance(feature_A , feature_C , N, fa_mean , fc_mean))

print("Covariance of feature_B & feature_A is =" ,Co_variance(feature_B , feature_A , N, fb_mean , fa_mean))
print("Covariance of feature_B & feature_B is =" ,Co_variance(feature_B , feature_B , N, fb_mean , fb_mean))
print("Covariance of feature_B & feature_C is =" ,Co_variance(feature_B , feature_C , N, fb_mean , fc_mean))

print("Covariance of feature_C & feature_A is =" ,Co_variance(feature_C , feature_A , N, fc_mean , fa_mean))
print("Covariance of feature_C & feature_B is =" ,Co_variance(feature_C , feature_B , N, fc_mean , fb_mean))
print("Covariance of feature_C & feature_c is =" ,Co_variance(feature_C , feature_C , N, fc_mean , fc_mean))
```

```python
def identity_matrix(n):     #n: the square size of the matrix
    IdM = n_matrix(3, 3)
    for i in range(n):
        IdM[i][i] = 1
 
    return IdM
```

```python
#Function for multiplying (3*3) * (3*3) metrices
def matrix_multiplication(mat_1 , mat_2, results):
  for i in range(len(mat_1)):
    # iterating by column of mat_1
    for j in range(len((mat_2[0]))):
 
        # iterating by rows of mat_2
        for k in range(len((mat_2))):
            results[i][j] += mat_1[i][k] * mat_2[k][j]
    return results
 
```
